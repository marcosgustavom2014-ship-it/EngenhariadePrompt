```python
def eh_primo(n):
  if n < 2:
    return False
  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
      return False
  return True
```

```python
# Testando a função eh_primo
print(f"É 7 primo? {eh_primo(7)}")      # Esperado: True
print(f"É 10 primo? {eh_primo(10)}")    # Esperado: False
print(f"É 2 primo? {eh_primo(2)}")      # Esperado: True
print(f"É 1 primo? {eh_primo(1)}")      # Esperado: False
print(f"É 0 primo? {eh_primo(0)}")      # Esperado: False
print(f"É 29 primo? {eh_primo(29)}")    # Esperado: True
```
