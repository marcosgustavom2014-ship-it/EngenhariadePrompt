Missao 01 - Automaçao de Arquivos

```python
import os


def organizar_jpg_pdf(caminho_pasta: str) -> None:
    if not os.path.exists(caminho_pasta):
        print("O caminho informado não existe.")
        return

    for item in os.listdir(caminho_pasta):
        caminho_item = os.path.join(caminho_pasta, item)

        if not os.path.isfile(caminho_item):
            continue

        _, extensao = os.path.splitext(item)
        extensao = extensao.lower()

        if extensao in (".jpg", ".jpeg"):
            pasta_destino = os.path.join(caminho_pasta, "Imagens")

        elif extensao == ".pdf":
            pasta_destino = os.path.join(caminho_pasta, "PDFs")

        else:
            continue

        os.makedirs(pasta_destino, exist_ok=True)
        os.rename(caminho_item, os.path.join(pasta_destino, item))

        print(f"{item} movido para {os.path.basename(pasta_destino)}/")

    print("Organização finalizada.")



```

```python
import requests

def consultar_cep(cep: str):
    url = f"https://viacep.com.br/ws/{cep}/json/"

    resposta = requests.get(url)
    dados = resposta.json()

    # Correção: verificar se a API retornou erro
    if "erro" in dados:
        print("❌ CEP não encontrado.")
        return

    print("📍 Endereço encontrado:")
    print("Logradouro:", dados.get("logradouro", "Não informado"))
    print("Bairro:", dados.get("bairro", "Não informado"))
    print("Cidade:", dados.get("localidade", "Não informado"))
    print("Estado:", dados.get("uf", "Não informado"))
```

Missao 03 - Sistema de Notificaçoes

```python
import time
import random
from datetime import datetime

# Configurações
LIMITE_CRITICO = 75
INTERVALO_VERIFICACAO = 5  # segundos
alerta_ativo = False

def ler_valor():
    """
    Simula a leitura de um sensor ou métrica crítica.
    """
    return random.randint(50, 90)

def enviar_alerta_console(valor):
    print(f"[ALERTA CRÍTICO] {datetime.now()} | Valor atingido: {valor}")

def enviar_email_simulado(valor):
    print("""
    -------------------------------
    EMAIL SIMULADO
    Para: admin@sistema.local
    Assunto: ALERTA CRÍTICO
    Mensagem:
    Valor crítico detectado: {}
    -------------------------------
    """.format(valor))

def monitorar():
    global alerta_ativo

    print("Monitoramento iniciado...\n")

    while True:
        valor_atual = ler_valor()
        print(f"Valor atual monitorado: {valor_atual}")

        if valor_atual >= LIMITE_CRITICO and not alerta_ativo:
            enviar_alerta_console(valor_atual)
            enviar_email_simulado(valor_atual)
            alerta_ativo = True

        elif valor_atual < LIMITE_CRITICO:
            alerta_ativo = False  # Reseta alerta quando normaliza

        time.sleep(INTERVALO_VERIFICACAO)

if __name__ == "__main__":
    monitorar()
```

```python
import time
import random
from datetime import datetime


class MonitorCritico:
    def __init__(self, limite_critico: int, intervalo_verificacao: int):
        self.limite_critico = limite_critico
        self.intervalo_verificacao = intervalo_verificacao
        self.alerta_disparado = False

    def ler_valor(self) -> int:
        """
        Simula a leitura de um valor monitorado.
        """
        return random.randint(50, 90)

    def condicao_critica(self, valor: int) -> bool:
        """
        Verifica se o valor ultrapassou o limite crítico.
        """
        return valor >= self.limite_critico

    def disparar_alerta(self, valor: int) -> None:
        """
        Dispara todos os alertas necessários.
        """
        self._alerta_console(valor)
        self._email_simulado(valor)

    def _alerta_console(self, valor: int) -> None:
        print(f"[ALERTA CRÍTICO] {datetime.now()} | Valor: {valor}")

    def _email_simulado(self, valor: int) -> None:
        print(f"""
        -------------------------------
        EMAIL SIMULADO
        Para: admin@sistema.local
        Assunto: ALERTA CRÍTICO
        Mensagem:
        Valor crítico detectado: {valor}
        -------------------------------
        """)

    def monitorar(self) -> None:
        print("Monitoramento iniciado...\n")

        while True:
            valor_atual = self.ler_valor()
            print(f"Valor atual: {valor_atual}")

            if self.condicao_critica(valor_atual):
                if not self.alerta_disparado:
                    self.disparar_alerta(valor_atual)
                    self.alerta_disparado = True
            else:
                self.alerta_disparado = False

            time.sleep(self.intervalo_verificacao)


if __name__ == "__main__":
    monitor = MonitorCritico(
        limite_critico=75,
        intervalo_verificacao=5
    )
    monitor.monitorar()
```
