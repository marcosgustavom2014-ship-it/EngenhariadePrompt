Criar um sistema de Python para verificar temperatura

```python
'''
Crie um script simples em python para verificar
a temperatura de um sistema. Se for maior que 30
exiba um alerta de superaquecimento.
'''

# Solicita ao usuário que insira a temperatura atual do sistema.
# O valor inserido é convertido para um número inteiro.
temperatura_atual = int(input("Digite a temperatura atual do sistema: "))

# Imprime a temperatura atual do sistema usando concatenação de strings
print("Temperatura atual do sistema: " + str(temperatura_atual) + "°C")

# Verifica se a temperatura atual é maior que 30
if temperatura_atual > 30:
    # Se for maior, exibe um alerta de superaquecimento
    print("ALERTA: Superaquecimento detectado!")
else:
    # Caso contrário, informa que a temperatura está normal
    print("Temperatura normal.")
```

```python
import os
import shutil
import random

# --- Configurações --- #
# Diretório onde os arquivos bagunçados serão criados e depois organizados
source_dir = 'arquivos_baguncados'
# Diretório onde as pastas organizadas serão criadas
target_dir = 'arquivos_organizados'

# Tipos de extensões para os arquivos de teste
file_extensions = ['.pdf', '.jpg', '.png', '.doc', '.docx', '.txt', '.xlsx', '.mp3', '.mp4']

# --- Criar diretórios (se não existirem) --- #
if not os.path.exists(source_dir):
    os.makedirs(source_dir)
    print(f"Diretório '{source_dir}' criado.")

if not os.path.exists(target_dir):
    os.makedirs(target_dir)
    print(f"Diretório '{target_dir}' criado.")

# --- Criar 15 arquivos de teste --- #
print("\nCriando 15 arquivos de teste...")
for i in range(1, 16):
    # Escolhe uma extensão aleatória
    ext = random.choice(file_extensions)
    # Gera um nome de arquivo
    file_name = f"arquivo_teste_{i}{ext}"
    file_path = os.path.join(source_dir, file_name)

    # Cria um arquivo vazio (ou com conteúdo simples para ser mais real)
    with open(file_path, 'w') as f:
        f.write(f"Este é o conteúdo do {file_name}\n")
    print(f"  - '{file_name}' criado em '{source_dir}'")

print("\nArquivos de teste criados com sucesso!")

# --- Organizar arquivos --- #
print("\nIniciando a organização dos arquivos...")

# Lista todos os arquivos no diretório de origem
files_to_organize = [f for f in os.listdir(source_dir) if os.path.isfile(os.path.join(source_dir, f))]

if not files_to_organize:
    print(f"Nenhum arquivo encontrado em '{source_dir}' para organizar.")
else:
    for filename in files_to_organize:
        # Constrói o caminho completo do arquivo de origem
        source_file_path = os.path.join(source_dir, filename)

        # Extrai a extensão do arquivo (ex: '.pdf', '.jpg')
        _, file_extension = os.path.splitext(filename)

        # Remove o ponto da extensão para usar como nome da pasta
        folder_name = file_extension[1:].lower() if file_extension else 'outros'

        # Constrói o caminho completo da pasta de destino para esta extensão
        destination_folder_path = os.path.join(target_dir, folder_name)

        # Cria a pasta de destino se ela não existir
        if not os.path.exists(destination_folder_path):
            os.makedirs(destination_folder_path)
            print(f"  - Pasta '{folder_name}' criada em '{target_dir}'.")

        # Constrói o caminho completo do arquivo no destino
        destination_file_path = os.path.join(destination_folder_path, filename)

        # Move o arquivo para a pasta de destino
        shutil.move(source_file_path, destination_file_path)
        print(f"  - '{filename}' movido para '{destination_folder_path}'.")

print("\nOrganização de arquivos concluída!")
print(f"Verifique os diretórios '{source_dir}' (agora vazio ou com menos arquivos) e '{target_dir}' para ver o resultado.")

# Opcional: Para limpar os diretórios criados após a execução, descomente as linhas abaixo.
# print("\nLimpando diretórios criados...")
# shutil.rmtree(source_dir)
# shutil.rmtree(target_dir)
# print("Diretórios de teste e organizados removidos.")
```
